```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
original_df = pd.read_csv("eng2_modified.csv")
augmented_df = pd.read_csv("eng_augmented.csv")

# Calculate emotion distributions
original_dist = original_df[['Anger', 'Fear', 'Joy', 'Sadness', 'Surprise']].mean()
augmented_dist = augmented_df[['Anger', 'Fear', 'Joy', 'Sadness', 'Surprise']].mean()

# Plot the distributions
df_dist = pd.DataFrame({'Original': original_dist, 'Augmented': augmented_dist})
df_dist.plot(kind='bar', figsize=(10,6))
plt.title('Emotion Distribution Comparison')
plt.ylabel('Proportion')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "eng_augmented.csv"
data = pd.read_csv(file_path)

# Summing the occurrences of each emotion
emotion_counts = data[['Anger', 'Fear', 'Joy', 'Sadness', 'Surprise']].sum()

# Plotting the bar graph
plt.figure(figsize=(8, 5))
bars = plt.bar(emotion_counts.index, emotion_counts.values)

# Adding text on top of each bar
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # X-coordinate
        bar.get_height(),                  # Y-coordinate
        f'{int(bar.get_height())}',        # Text to display
        ha='center', va='bottom'           # Alignment
    )

plt.title('Emotion Data Balance')
plt.xlabel('Emotions')
plt.ylabel('Number of Points')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Remove the points with only 'Fear' or 'Fear' and 'Joy' from row 2900 to the last
import pandas as pd

# Load the dataset
file_path = "filtered_emotion_data.csv"
balanced_data = pd.read_csv(file_path)
filtered_data = balanced_data.iloc[1800:].drop(
    balanced_data.iloc[1800:][
        (balanced_data['Fear'] == 1) &
        ((balanced_data[['Anger', 'Sadness', 'Surprise']].sum(axis=1) == 0) & (balanced_data['Joy'] <= 1))
    ].index
)

# Combine the filtered portion with the unchanged data from rows 0 to 2899
balanced_data_updated = pd.concat([balanced_data.iloc[:1800], filtered_data], ignore_index=True)

# Save the updated dataset to a new file
balanced_data_updated.to_csv("filtered_emotion_data2.csv", index=False)
```

```python
import pandas as pd

# Load the filtered CSV file
file_path = 'filtered_emotion_data4.csv'
data = pd.read_csv(file_path)

# Find rows where only 'Anger' and 'Fear' are 1, and others are 0
fear_anger_rows = data[
    (data['Anger'] == 0.0) & 
    (data['Fear'] == 1.0) & 
    (data['Joy'] == 0.0) & 
    (data['Sadness'] == 0.0) & 
    (data['Surprise'] == 1.0)
]

# Identify the indices of the last 30 rows matching the condition
indices_to_remove = fear_anger_rows.index[-20:]

# Remove these rows from the dataset
filtered_data = data.drop(indices_to_remove)

# Save the updated dataset to a new CSV file
filtered_data.to_csv('filtered_emotion_data4.csv', index=False)
```
